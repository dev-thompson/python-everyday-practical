```python
# Student Records (2D List)

students = [
    ["John", "Computer Science", 85],
    ["Mary", "Mathematics", 92],
    ["David", "Physics", 78],
    ["Grace", "Chemistry", 88],
    ["Paul", "Engineering", 95]
]

print("All Students:\n")

for student in students:
    print("Name:", student[0])
    print("Department:", student[1])
    print("Score:", student[2])
    print()

# Find highest score student
highest = students[0]

for student in students:
    if student[2] > highest[2]:
        highest = student

print("Highest Score Student")
print("Name:", highest[0])
print("Department:", highest[1])
print("Score:", highest[2])
```
